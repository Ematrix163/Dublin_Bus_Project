```python
import pandas as pd
import os, sys


file_dir = os.path.dirname(os.path.abspath("__file__"))
parent_dir = os.path.dirname(file_dir)
newPath = os.path.join(parent_dir, 'data/db_historic/rt_trips_2017_I_DB.txt')

# read trips 2016 .txt file into dataframe16
trips17 = pd.read_csv(newPath, delimiter=';')




```

```python
# THIS cell removes all rows where suppressed is equal to 1, then counts to check there are no rows with this values left
suppressed_is_1 = trips17['suppressed']==1  
is_suppressed = trips17[suppressed_is_1]
print("Number of rows in trips16 where suppressed column has value 1 (before they are removed): ", is_suppressed.shape[0])

trips17= trips17[trips17.suppressed != 1]
    
suppressed_is_1 = trips17['suppressed']==1  
is_suppressed = trips17[suppressed_is_1]
print("Number of rows in trips16 where suppressed column has value 1: ", is_suppressed.shape[0])
print()

    
    
```

```python
# THIS CELL WILL DROP the column 'datasource'

trips17 = trips17.drop('datasource', axis=1)
trips17.head()
```

```python
# THIS CELL WILL DROP the column 'basin'

trips17 = trips17.drop('basin', axis=1)
trips17.head()
```

```python
# THIS CELL WILL DROP the column 'tenderlot'

trips17 = trips17.drop('tenderlot', axis=1)
trips17.head()
```

```python
# THIS CELL WILL DROP the column 'justificationid'

trips17 = trips17.drop('justificationid', axis=1)
trips17.head()
```

```python
# THIS CELL WILL DROP the column 'lastupdate'

trips17 = trips17.drop('lastupdate', axis=1)
trips17.head()
```

```python
# THIS CELL WILL DROP the column 'note'

trips17 = trips17.drop('note', axis=1)
trips17.head()
```

```python
# THIS CELL WILL change datatype of dayofservice from string to datetime format
trips17['dayofservice'] =  pd.to_datetime(trips17['dayofservice'], dayfirst=True)
```
