```python
import pandas as pd
import os, sys



file_dir = os.path.dirname(os.path.abspath("__file__"))
parent_dir = os.path.dirname(file_dir)
newPath = os.path.join(parent_dir, 'data/db_historic/rt_leavetimes_2017_I_DB.txt')

# read leavetimes 2016 .txt file into csv leave16 (just first 100)
leave17 = pd.read_csv(newPath, delimiter=';', chunksize=100)


# just put 10 in a dataframe for the moment...
leave17chunk=leave17.get_chunk(10)
leave17chunk
```

```python
# use chunk size

chunk_size = 1000000
count=1
suppressedIsOneCount=0

# load the big file in smaller chunks
for leave_chunk in pd.read_csv(newPath, delimiter=';', chunksize=chunk_size):
    print("--------------------")
    print("Chunk number: ", count)
    print()
    print("datasource has value null in this chunk:", leave_chunk['datasource'].isnull().sum())
    print("dayofservice has value null in this chunk:", leave_chunk['dayofservice'].isnull().sum())
    print("tripid has value null in this chunk:", leave_chunk['tripid'].isnull().sum())
    print("progrnumber has value null in this chunk:", leave_chunk['progrnumber'].isnull().sum())
    print("stoppointid has value null in this chunk:", leave_chunk['stoppointid'].isnull().sum())
    print("plannedtime_arr has value null in this chunk:", leave_chunk['plannedtime_arr'].isnull().sum())
    print("plannedtime_dep has value null in this chunk:", leave_chunk['plannedtime_dep'].isnull().sum())
    print("actualtime_arr has value null in this chunk:", leave_chunk['actualtime_arr'].isnull().sum())
    print("actualtime_dep has value null in this chunk:", leave_chunk['actualtime_dep'].isnull().sum())
    print("vehicleid has value null in this chunk:", leave_chunk['vehicleid'].isnull().sum())
    print("passengers has value null in this chunk:", leave_chunk['passengers'].isnull().sum())
    print("passengersin has value null in this chunk:", leave_chunk['passengersin'].isnull().sum())
    print("passengersout has value null in this chunk:", leave_chunk['passengersout'].isnull().sum())
    print("distance has value null in this chunk:", leave_chunk['distance'].isnull().sum())
    print("suppressed has value null in this chunk:", leave_chunk['suppressed'].isnull().sum())
    print("justificationid has value null in this chunk:", leave_chunk['justificationid'].isnull().sum())
    print("lastupdate has value null in this chunk:", leave_chunk['lastupdate'].isnull().sum())
    print("note has value null in this chunk:", leave_chunk['note'].isnull().sum())
    print()
    # check for duplicate rows
    print('Duplicate rows:', leave_chunk.duplicated()[leave_chunk.duplicated() == True].shape[0])
    # Check for duplicate columns
    print('Duplicate columns:',leave_chunk.columns.size - leave_chunk.columns.unique().size)
    print()
    print("FINISHED A CHUNK")
    print()
    print()
    suppressed_is_1 = leave_chunk['suppressed']==1  
    is_suppressed = leave_chunk[suppressed_is_1]
    print("Number of rows in this chunk where suppressed column has value 1: ", is_suppressed.shape[0])
    print()
    #leave_chunk.to_csv('leave16_prelim.csv', index=True, header=False, mode='a')   
    suppressedIsOneCount=suppressedIsOneCount+is_suppressed.shape[0]
    count=count+1
    #if count==4:
     #   break
        
        
print()
print("Total no. of suppressed = 1 values in leave17 txt file is: ", suppressedIsOneCount)
   
    
print()
print("CELL FINISHED")
```

CONCLUSIONS:
- There are NO duplicate rows or columns in whole of leave2017 txt file
- Passengers, Passengersin and Passengersout are all null in this file. (as they are in leave16 txt file). So the three columns will be dropped in the csv.
- ALL 'distance' columns are null/NaN, meaning it should candidate to be dropped, (all distance values were also null in the leave2016 txt file
- no. of suppressed=1 values are: 404,511. These are to be dropped in final cleanse
- although there are some non-null entries in 'justificationid', the feature records 'fault codes' to do with why a journey was suppressed, so is not relevant to developing our model, so we will be dropping this column.
- 'lastupdate' is not relevant to our model either, as it does not relate to what day a journey took place, only when the data was updated. So we will drop this column completely also.
- ALL 'note' columns are null/NaN, (which is also true in the leave 2016 txt file), so we will be dropping this feature.
- We also have previously discoverd that the feature 'datasource' has a cardinality of one: ie, it only has one value, DB, for the entire dataset. Therefore we will drop this feature.
- The Vehicleid feature will also be dropped as it is not relevant to developing our model.
