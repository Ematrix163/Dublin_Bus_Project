## Finding rows that are duplicates in order to find a key for schema

### 2016 Data

```python
df16.dtypes
```

```python
df16.shape
```

```python
#finding duplicate rows
df16["is_duplicate"]= df16.duplicated()
df16.head()
```

```python
df16 = df16[df16.is_duplicate != True]
```

```python
df16.shape
```

```python
df16.dt_iso.value_counts()
```

```python
duplicate = df16['dt_iso'] == "2016-12-06 03:00:00 +0000 UTC"
df16[duplicate]
```

```python
#finding rows where the data is the same
df16["is_duplicate"]= df16.duplicated(subset = 'dt')
duplicate = df16['dt_iso'] == "2016-12-06 03:00:00 +0000 UTC"
df16[duplicate]
```

```python
#deleting rows where the date is the same
df16 = df16[df16.is_duplicate != True]
```

```python
df16.shape
```

## Changing the date format in holidays csv file

```python
import pandas as pd
import os, sys


#reading in full historic weather data
file_dir = os.path.dirname(os.path.abspath("__file__"))
parent_dir = os.path.dirname(file_dir)
newPath = os.path.join(parent_dir, 'data/holidays.csv')


holidays = pd.read_csv(newPath, delimiter=',')
```

```python
holidays.head(5)
```

```python
holidays['date'] =  pd.to_datetime(holidays['date'], dayfirst=True)
```

```python
holidays.head(5)
```

```python
holidays.to_csv('holidays.csv', index=False, header= True)
```
