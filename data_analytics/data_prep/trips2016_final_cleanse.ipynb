```python
import pandas as pd
import os, sys


file_dir = os.path.dirname(os.path.abspath("__file__"))
parent_dir = os.path.dirname(file_dir)
newPath = os.path.join(parent_dir, 'data/db_historic/rt_trips_2016_I_DB.txt')

# read trips 2016 .txt file into dataframe16
trips16 = pd.read_csv(newPath, delimiter=';')

#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)


```

```python
# THIS cell removes all rows where suppressed is equal to 1, then counts to check there are no rows with this values left


trips16= trips16[trips16.suppressed != 1]
    
suppressed_is_1 = trips16['suppressed']==1  
is_suppressed = trips16[suppressed_is_1]
print("Number of rows in trips16 where suppressed column has value 1: ", is_suppressed.shape[0])
print()

    
    
```

```python
# THIS CELL WILL DROP the column 'datasource'

trips16 = trips16.drop('datasource', axis=1)
trips16.head()
```

```python
# THIS CELL WILL DROP the column 'basin'

trips16 = trips16.drop('basin', axis=1)
trips16.head()
```

```python
# THIS CELL WILL DROP the column 'tenderlot'

trips16 = trips16.drop('tenderlot', axis=1)
trips16.head()
```

```python
# THIS CELL WILL DROP the column 'justificationid'

trips16 = trips16.drop('justificationid', axis=1)
trips16.head()
```

```python
# THIS CELL WILL DROP the column 'lastupdate'

trips16 = trips16.drop('lastupdate', axis=1)
trips16.head()
```

```python
# THIS CELL WILL DROP the column 'note'

trips16 = trips16.drop('note', axis=1)
trips16.head()
```

```python
# THIS CELL WILL change datatype of dayofservice from string to datetime format
trips16['dayofservice'] =  pd.to_datetime(trips16['dayofservice'], dayfirst=True)
```

```python
trips16.head()
```

```python
#THIS CELL WILL CREATE THE CSV

#CREATE CSV of trips16
trips16.to_csv('trips16.csv', index=False, header=True)  
```
