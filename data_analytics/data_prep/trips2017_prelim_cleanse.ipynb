'Suppressed' is the feature that might be most relevant in data cleaning. Of the 929507 rows in the entire dataframe, 927797 are NaN for suppressed. With just 1710 having actual values (1.0 or 0.0). The frequency tables below show that there are 1409 rows with 1.0 and 301 with 0.0. (frequency tables don't seem to record NaN values).

Here is what is said in the NTA pdf about the feature 'suppressed':
"If the trip is planned and achieved, the record will contain both the planned and actual data. If the
tripped is planned and suppressed the record will contain only the planned data and the flag
“Suppressed” will be checked.
If the trips is partially suppressed the flag “Suppressed” will not be checked. The set of the leave
times suppressed is detailed in table RT_LeaveTimes"

It also says, in explaining suppressed: "The whole trip has been suppressed (0 =achieved, 1 = suppressed)"

So rows with the value 1 for suppressed must be dropped - if they only have data for planned times, not actual times, then they will be of no use in training the model. So this will be dropped when the trips2016 csv is created.

CONCLUSIONS:
- There are NO duplicate rows or columns in whole of trips 2017 txt file
- datasource column can be dropped as it just has a single value, DB.
- Basin can be dropped (it just has a single value, BasDef, which is of no relevance to our model).
- For the Tenderlot feature, all entries are NaN. So we will drop it.
- although there are some entries in 'justificationid', as the feature records 'fault codes' to do with why a journey was suppressed - this is not relevant to developing our model, so we will be dropping this column. 
- 'note' is not relevant to our model either, so will be dropped 
- 'lastupdate' is not relevant to our model either, as it does not relate to what day a journey took place, only when the data was updated. So we will drop this column completely also.
- We will also be dropping all rows in the suppressed column where suppressed=1
