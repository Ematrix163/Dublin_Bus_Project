```python
file_dir = os.path.dirname(os.path.abspath("__file__"))
parent_dir = os.path.dirname(file_dir)
newPath = os.path.join(parent_dir, '/home/student/data/weather/2017weatherClean.csv')


dfW17 = pd.read_csv(newPath, delimiter=',')
```

```python
#Only keeping the columns that we want
dfW17  = dfW17[['dt','temp','pressure','humidity','wind_speed','weather_description','clouds_all','wind_deg']]
```

```python
#We want to look at only one direction
df = df.loc[(df['Direction'] == 1)]
```

```python
df = df[['DayOfService', 'TripID', 'ProgrNumber', 'StopPointID', 'PlannedTime_Arr_leave', 'ActualTime_Arr_leave', 'RouteID']]
```

```python
import numpy as np
import datetime as dt
month = pd.to_datetime(df['DayOfService']).dt.month
month = pd.DataFrame(data={'month':month})
month['month'] = month['month'].astype(np.int64)


day = pd.to_datetime(df['DayOfService']).dt.day
day = pd.DataFrame(data={'day':day})
day['day'] = day['day'].astype(np.int64)

dayofweek = pd.to_datetime(df['DayOfService']).dt.dayofweek
dayofweek = pd.DataFrame(data={'dayofweek':dayofweek})
dayofweek['dayofweek'] = dayofweek['dayofweek'].astype(np.int64)

dt = pd.DatetimeIndex(df['DayOfService']).astype(np.int64)/1000000000 + df['ActualTime_Arr_leave']
dt = pd.DataFrame(data={'dt':dt})
dt['dt'] = dt['dt'].astype(np.int64)
```

```python
df = pd.concat([dt, month, day, dayofweek, df], axis =1)
```

```python
df.head()
```

```python
df['delay'] = df['ActualTime_Arr_leave'] - df['PlannedTime_Arr_leave']
```
